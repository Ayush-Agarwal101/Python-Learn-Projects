# Random Distribution Graphs

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

**Quadratic function**

```python
x = np.arange(-50,50)
y = np.arange(-1,1000)
```

```python
y = x * x
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Quadratic function graph')
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Quadratic function graph in scattered form')
plt.scatter(x,y)
```

**Polynomial function**

```python
y = x * x * x * x * x
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Polynomial function graph (x^5) ')
```

```python
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Polynomial function graph (x^5) in scattered form')
plt.scatter(x,y)
```

**Bar Graph**

```python
x1 = np.array([2020, 2021, 2022, 2023, 2024])
y1 = np.array([30, 25, 45, 15, 20])
plt.bar(x1,y1, color = 'hotpink')
plt.xlabel('Year')
plt.ylabel('Annual Return (in lakhs) ')
plt.title('Annual Return of XYZ Company Per Year')
plt.show()
```

**Pie Chart**

```python
plt.pie(y1, labels = x1)
plt.title('Annual Return of XYZ Company Per Year (in lakhs)')
plt.show()
```

**Histogram Uniform Distribution**

```python
arr = np.random.uniform(0,5,2000)
plt.hist(arr, 100)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

**Normal Distribution**

```python
import seaborn as sns
```

```python
arr1 = np.random.normal(size = 20000)
sns.displot(arr1, kind = 'kde')
plt.title('Normal Distribution Curve')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

**Binomial Distribution**

```python
arr2 = np.random.binomial(n= 100,p = 0.5, size = 20000)
sns.displot(arr2, kind = 'kde')
plt.title('Binomial Distribution Curve')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

**Exponential Distribution**

```python
arr3 = np.random.exponential(size = 20000)
sns.displot(arr3, kind = 'kde')
plt.title('Exponential Distribution Curve')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
arr4 = np.random.uniform(size=20000)
sns.displot(arr4, kind = 'kde')
plt.title('Uniform Distribution Curve')
plt.xlabel('x axis')
plt.ylabel('y axis')
```
