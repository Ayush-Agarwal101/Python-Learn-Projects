# Seaborn Assignment

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
x1 = sns.load_dataset('diamonds').head(25)
x1
```

```python
sns.lineplot( x = "carat", y = "price", data = x1, hue = "cut", style = "cut")
plt.grid()
plt.title("Diamonds")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
